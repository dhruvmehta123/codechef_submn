```python
import pandas as pd
import matplotlib.pyplot as plt



```

```python
df=pd.read_csv("sales_data.csv")
df
```

```python
#Task 1 (a)
plt.plot(df['ID'],df['Revenue'])
plt.xlabel('ID')
plt.ylabel('Revenue')
```

```python
#Task 1 (b)
plt.scatter(df['Revenue'],df['Quantity'])
plt.xlabel('Revenue')
plt.ylabel('Quantity')
```

```python
#task 1 (c)
product_category_revenue = df.groupby('Product')['Revenue'].sum()
plt.figure(figsize=(10, 6))
product_category_revenue.plot(kind='bar')
plt.title('Total Revenue for Each Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
```
